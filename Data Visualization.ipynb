## Renaming data set for concise messages

```python
s_data = pd.read_csv('Mental_Health_Survey_Feb_20_22.csv')
s_data
clean_data = s_data[['acha_12months_times_5', 'acha_12months_times_6','acha_12months_times_7', 'acha_depression',
                         'acha_services_1', 'acha_services_2', 'acha_services_3']]
new_column_names = {
    'acha_12months_times_5': 'Felt Depressed',
    'acha_12months_times_6': 'Considered Attempting Suicide',
    'acha_12months_times_7': 'Attempted Suicide',
    'acha_depression': 'Diagnosed with Depression',
    'acha_services_1': 'Diagnosed within 12 months',
    'acha_services_2': 'In therapy',
    'acha_services_3': 'Taking medications'
}
clean_data = clean_data.rename(columns=new_column_names)
clean_data = clean_data.drop([0, 1])
print(clean_data)
```

## Converting string answers to int for future analysis

```python
mapping1 = {
    'Never': 0,
    '1-2 times': 2,
    '3-4 times': 4,
    '5-6 times': 6,
    '7-8 times': 8,
    '9-10 times': 10,
    '11 or more times': 11
}

# Convert 'Felt Depressed' column to int values
clean_data['Felt Depressed'] = clean_data['Felt Depressed'].map(mapping1)
clean_data['Considered Attempting Suicide'] = clean_data['Considered Attempting Suicide'].map(mapping1)
clean_data['Attempted Suicide'] = clean_data['Attempted Suicide'].map(mapping1)

clean_data
```

```python
categories = ['Never', '1-2 times', '3-4 times', '5-6 times', '7-8 times', '9-10 times', '11 or more times']
# Create a new categorical column
clean_data['Considered Attempting Suicide'] = pd.cut(clean_data['Considered Attempting Suicide'], 
                                     bins=[-1, 0, 2, 4, 6, 8, 10, float('inf')], 
                                     labels=categories, 
                                     right=True)

# Plot the distribution of categorized people who considered attempting suicide
plt.figure(figsize=(8, 6))
clean_data['Considered Attempting Suicide'].value_counts().sort_index().plot(kind='bar', color='skyblue')

plt.title('Distribution of People Who Considered Attempting Suicide')
plt.xlabel('Frequency of Considering Suicide')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

## Visualization of people diagnoised with Depression

```python
# Count the number of people diagnosed with depression who are in therapy or not
therapy_counts = clean_data.groupby('In therapy')['Diagnosed with Depression'].apply(lambda x: (x == 'Yes').sum()).reset_index()

# Plotting the graph
plt.figure(figsize=(8, 6))
plt.bar(therapy_counts['In therapy'], therapy_counts['Diagnosed with Depression'], color=['skyblue', 'salmon'])

plt.title('Number of People Diagnosed with Depression In Therapy vs Not In Therapy')
plt.xlabel('In Therapy')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

## Statistics

```python
# Count the number of people diagnosed with depression who are in therapy
num_in_therapy = (clean_data['Diagnosed with Depression'] == 'Yes').sum()

num_not_therapy = (clean_data['Diagnosed with Depression'] == 'No').sum()

# Count the total number of people diagnosed with depression
total_diagnosed = num_in_therapy + num_not_therapy

# Calculate the percentage in therapy
percentage_in_therapy = (num_in_therapy / total_diagnosed) * 100

print(f"The percentage of people diagnosed with depression who are in therapy is: {percentage_in_therapy:.2f}%")
```
